#### üîå Load the Stable Diffusion Pipeline

```python
from diffusers import StableDiffusionPipeline
import torch

# Detect the best device
if torch.backends.mps.is_available():
    device = "mps"
elif torch.cuda.is_available():
    device = "cuda"
else:
    device = "cpu"

# Choose appropriate dtype
dtype = torch.float32 if device == "cpu" else torch.float16

# Load the Stable Diffusion pipeline
pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=dtype,
    use_safetensors=True,
)

# Move pipeline to the best device
pipe = pipe.to(device)

print(f"Model loaded and ready on: {device}")
```

#### üí¨ -> üèûÔ∏è Generate Image from Text

```python
prompt = "a train with wings"
image = pipe(prompt).images[0]

image.show()
```
